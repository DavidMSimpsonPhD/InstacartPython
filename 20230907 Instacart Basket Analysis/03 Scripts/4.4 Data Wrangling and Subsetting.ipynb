#### Calculating orders by hour percentages for final report

```python
# Calculate the total count of occurrences
total_count = hour_counts.sum()
```

```python
# Calculate the percentage of total occurrences for each hour
hour_percentages = (hour_counts / total_count) * 100
```

```python
# Create a DataFrame with numerical indices and percentages
hour_df = pd.DataFrame({'Hour of the Day': hour_percentages.index, 'Percentage of Total': hour_percentages.values})
```

```python
# Sort the DataFrame by the hour of the day
hour_df = hour_df.sort_values(by='Hour of the Day')
```

```python
# Reset the index to create a numerical index order
hour_df = hour_df.reset_index(drop=True)
```

```python
# Display the resulting DataFrame
print(hour_df)
```
